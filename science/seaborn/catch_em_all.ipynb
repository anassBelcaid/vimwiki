## Importing the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading the dataset
df = pd.read_csv('Pokemon.csv', index_col=0,encoding='latin1')
df.head()
```

## Seaborn plotting functions

One of the advantages of `seaborn` is the ease and diversity of plotting functions. For example if we want a **scatter plot**. we use `lmplot`

```python
sns.lmplot(x='Attack', y='Defense', data=df)

#alternative
```

Here is a useful tweaks

- `fit_reg= False` : to remove the regression line
- `hue = 'Stage`   : color by another variable, and allow to express a third information

```python
sns.lmplot(x='Attack', y = 'Defense', hue='Stage', data=df, fit_reg=False)
```

## Customize matplotlib

```python
sns.lmplot(x='Attack', y = 'Defense', data=df,
          fit_reg = False,
           hue = 'Stage'
          )

plt.ylim(0,None)
plt.xlim(0,None)
```

## Role of pandas

```python
sns.boxplot(data=df)
```

```python
stats_df= df.drop(['Total', 'Stage', 'Legendary'], axis=1)
sns.boxplot(data=stats_df)
```

## Seaborn themes

```python
sns.set_style('whitegrid')
fig, axs = plt.subplots(1,1, figsize=(8,6))
sns.violinplot(x='Type 1', y='Attack', data =df, ax=axs)
```

```python
sns.swarmplot(x='Type 1', y='Attack', data=df)
```

## Overlaping plots

```python
plt.figure(figsize=(10,6))

#first plot
sns.violinplot(x='Type 1', y = 'Attack', data=df, inner=None)

sns.swarmplot(x='Type 1', y='Attack', data=df, color='k', alpha=0.7)
plt.title('Attack by type')
```

## Heatmap

```python
corr = stats_df.corr()
```

```python
sns.heatmap(corr,cmap='RdBu')
```

```python
sns.distplot(df.Attack)
```

## Bar plot

```python
sns.countplot(x='Type 1', data=df)
plt.xticks(rotation=-45)
```

## Factor plot

```python
g = sns.factorplot(x='Type 1', y = 'Attack',
                  data=df, hue='Stage',
                  col='Stage', kind='swarm')
```

## Density plot

```python
sns.kdeplot(df.Attack, df.Defense)
```

## Joint distribution

```python
sns.jointplot(x='Attack', y='Defense', data=df)
```
